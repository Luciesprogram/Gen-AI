```python
#setup prompts - specific to stableLM
from llama_index.core import PromptTemplate

system_prompt = """<|SYSTEM|>#You are a Q/A assistant. Your task is to answer the questions as accuarately as possible and instructions based on the reterived context."""

# This will wrap the default prompts that are internal to llama-index
query_wrapper_prompt = PromptTemplate("<|USRE|>{query_str}<|ASSISTANT|>")
```

https://github.com/run-llama/llama_index/blob/main/llama-index-integrations/llms/llama-index-llms-huggingface/llama_index/llms/huggingface/base.py

```python
import torch
from transformers import BitsAndBytesConfig

quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
)
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.7, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="mistralai/Mistral-7B-Instruct-v0.2",
    model_name="mistralai/Mistral-7B-Instruct-v0.2",
    device_map="auto",
    stopping_ids=[50278, 50279, 50277, 1, 0],
    tokenizer_kwargs={"max_length": 4096},
    # Pass the quantization config here
    model_kwargs={
        "quantization_config": quantization_config,
        "torch_dtype": torch.float16
    }
)
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# Add device="cpu" to keep the GPU clear for the Mistral model
embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-mpnet-base-v2",
    device="cpu"
)
```

```python
from llama_index.core.indices import VectorStoreIndex
from llama_index.core import Settings

Settings.chunk_size=1024
Settings.llm=llm
Settings.embed_model=embed_model
```

```python
index = VectorStoreIndex.from_documents(documents, llm=llm, embed_model=embed_model)
```

```python
query_engine = index.as_query_engine()
```

```python
query_engine.query("what is attention?")
```

```python
from IPython.display import display,Markdown
import textwrap
```

```python
def gen(text):
    # 1. Get the Response object
    response = query_engine.query(text)

    # 2. Convert response to string using str()
    # This extracts only the actual answer text
    response_text = str(response)
    # print(response_text)

    # 3. Apply indenting to the string
    indented_text = textwrap.indent(response_text, '> ', predicate=lambda _: True)

    return Markdown(indented_text)
```

```python
gen("what is attention")
```

```python
gen("how attention is different from rnn and lstm")
```
